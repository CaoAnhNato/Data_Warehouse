# Import Library

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.drop('Row ID',axis=1,inplace=True)  #Dropping the Row ID column
```

# Định Dạng Kiểu Dữ Liệu Ngày Tháng

```python
df['Order Date']=pd.to_datetime(df['Order Date'], format='%d/%m/%Y') #convert Order dates to pandas datetime format
df['Ship Date']=pd.to_datetime(df['Ship Date'], format='%d/%m/%Y') #convert Ship Date to pandas datetime format
df
```

```python
#Checking NULL values
df.isnull().sum()
```

# Điền postal code cho thành phố bị thiếu

```python
df.loc[df['City'] == 'Burlington', 'Postal Code'] = df.loc[df['City'] == 'Burlington', 'Postal Code'].fillna(5401)
df.isnull().sum()
```

```python
df.info()
```

 # Đổi tên một số cột

```python
df.rename(columns={'Order ID':'Order_ID'},inplace=True)
df.rename(columns={'Postal Code':'Postal_Code'},inplace=True)
df.rename(columns={'Order Date':'Order_Date'},inplace=True)
df.rename(columns={'Ship Date':'Ship_Date'},inplace=True)
df.rename(columns={'Ship Mode':'Ship_Mode'},inplace=True)
df.rename(columns={'Customer ID':'Customer_ID'},inplace=True)
df.rename(columns={'Customer Name':'Customer_Name'},inplace=True)
df.rename(columns={'Product ID':'Product_ID'},inplace=True)
df.rename(columns={'Product Name':'Product_Name'},inplace=True)
df.rename(columns={'Sub-Category':'Sub_Category'},inplace=True)
df.info()
```

```python
df.head()
```

# Loại bỏ khoảng trắng

```python
# Remove whitespace in all the dataframe columns
df.columns = df.columns.str.strip()
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

# Loại bỏ kí tự ';' ở các cột object type

```python
df.columns = df.columns.str.strip()
df = df.apply(lambda x: x.str.replace(';','') if x.dtype == "object" else x)
```

# Chuyển toàn bộ dữ liệu cột Sales về 1 form nhất định

```python
def validate_sales_value(value):
    """
    Validate giá trị Sales và đưa về phạm vi hợp lý
    """
    # Ngưỡng min/max có thể điều chỉnh theo yêu cầu
    MIN_VALID = 0.0
    MAX_VALID = 1000000.0  # 1 triệu
    
    try:
        val = float(value)
        if pd.isna(val):
            return np.nan
            
        # Kiểm tra phạm vi hợp lệ
        if val < MIN_VALID or val > MAX_VALID:
            return np.nan  # hoặc giá trị mặc định
            
        return round(val, 2)  # Làm tròn 2 chữ số thập phân
    except (ValueError, TypeError):
        return np.nan

# Áp dụng hàm validate
df['Sales'] = df['Sales'].apply(validate_sales_value)
```

# Loại Bỏ Các Cột Không Cần Thiết

```python
df.drop(columns=['Ship_Date','Postal_Code'],axis=1,inplace=True)  #Dropping the Row ID column
df.head()
```
